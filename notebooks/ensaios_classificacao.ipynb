## KNN Classifier

### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics as mt
```

### Sobre os dados de treinamento

```python
#Importando os dados

x_train = pd.read_csv('../../dataset/Classificacao/Treinamento/X_training.csv')
y_train = pd.read_csv('../../dataset/Classificacao/Treinamento/y_training.csv')
```

```python
# Converte as variáveis x_train e y_train para arrays Numpy
x_train = x_train.to_numpy() 
y_train = y_train.to_numpy() 
```

```python
#Escolhendo o melhor parâmetro

acc = []

for k in range(3,10,2):
    
    knn_classifier = KNeighborsClassifier(n_neighbors=k)

    knn_classifier.fit(x_train,y_train)

    y_pred = knn_classifier.predict(x_train)

    acc.append(round(mt.accuracy_score(y_train,y_pred)*100,2))
```

```python
# Gerando o gráfico
plt.plot(range(3, 10, 2), acc, marker='o', color='b')
plt.title('Acurácia vs Número de Vizinhos (k)')
plt.xlabel('Número de Vizinhos (k)')
plt.ylabel('Acurácia (%)')
plt.grid(True)
plt.show()
```

```python
#Treinamento com melhor parâmetro

k = 3

knn_classifier = KNeighborsClassifier(n_neighbors=k)

knn_classifier.fit(x_train, y_train)

y_pred = knn_classifier.predict(x_train)
```

```python
#Métricas de performance
knn_train_acc = mt.accuracy_score(y_train,y_pred)
knn_train_precision = mt.precision_score(y_train,y_pred)
knn_train_recall =mt.recall_score(y_train,y_pred)
knn_train_f1 = round(mt.f1_score(y_train, y_pred, average='binary') * 100, 2) 

print(
    f"""A performance do algoritmo KNN em classificação para dados apenas de treinamento é:
    Accuracy: {knn_train_acc * 100:.2f}%
    Precision: {knn_train_precision}
    Recall: {knn_train_recall}
    F1 Score: {knn_train_f1}%"""
)
```

## Decision Tree Classifier

### Imports

```python
import pandas as pd
import numpy as np

from sklearn import metrics as mt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

### Sobre os dados de Treinamento

```python
#Importando os dados

x_train = pd.read_csv('../../dataset/Classificacao/Treinamento/X_training.csv')
y_train = pd.read_csv('../../dataset/Classificacao/Treinamento/y_training.csv')
```

```python
#Busca de melhores parâmetros busca em grade (Grid Search)

# Defina o modelo
modelo = DecisionTreeClassifier()

# Parâmetros a serem testados
param_grid = {
    'max_depth': [3, 5, 10, 20]
}

# Busca em grade com validação cruzada, cv=cross validation
grid_search = GridSearchCV(estimator=modelo, param_grid=param_grid, cv=5, scoring='accuracy')

# Treine o modelo
grid_search.fit(x_train, y_train)

# Melhor combinação de parâmetros
print("Melhores parâmetros:", grid_search.best_params_)
print("Melhor desempenho:", grid_search.best_score_)

best_max_depth = grid_search.best_params_['max_depth']
```

```python
#Treinamento do algoritmo com os melhores parâmetros

# Inicializando o classificador de árvore de decisão
tree_clf = DecisionTreeClassifier(max_depth=best_max_depth, random_state=42)

# Treinando a árvore de decisão
tree_clf.fit(x_train, y_train)

# Fazendo previsões
y_pred = tree_clf.predict(x_train)
```

```python
# Calculando as métricas de performance
decision_tree_classifier_train_acc = mt.accuracy_score(y_train, y_pred)
decision_tree_classifier_train_precision = mt.precision_score(y_train, y_pred, average='weighted')
decision_tree_classifier_train_recall = mt.recall_score(y_train, y_pred, average='weighted')
decision_tree_classifier_train_f1 = mt.f1_score(y_train, y_pred, average='weighted')

# Exibindo os resultados
print("Métricas de Performance:")
print(f"Accuracy: {decision_tree_classifier_train_acc:.2f}")
print(f"Precision: {decision_tree_classifier_train_precision:.2f}")
print(f"Recall: {decision_tree_classifier_train_recall:.2f}")
print(f"F1-Score: {decision_tree_classifier_train_f1:.2f}")
```

## Random Forest Classifier

### Imports

```python
import pandas as pd
import numpy as np

from sklearn import metrics as mt
from sklearn.utils import resample
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

### Sobre os dados de Treinamento

```python
#Importando os dados de treino
x_train = pd.read_csv('../../dataset/Classificacao/Treinamento/X_training.csv')
y_train = pd.read_csv('../../dataset/Classificacao/Treinamento/y_training.csv')
```

```python
#Reduzindo o conjunto de dados de forma aleatória

# Amostra aleatória sem reposição
x_sample, y_sample = resample(
    x_train, y_train,
    n_samples=1000,       
    random_state=42       
)
```

```python

# Parâmetros para busca aleatória
param_dist = {
    'n_estimators': [int(x) for x in np.linspace(start=100, stop=1000, num=10)],
    'max_depth': [int(x) for x in np.linspace(10, 110, num=11)] + [None]
}

# Instancia o modelo RandomForestClassifier
rf = RandomForestClassifier()

# RandomizedSearchCV para busca aleatória
random_search = RandomizedSearchCV(
    estimator=rf,                     
    param_distributions=param_dist,   
    n_iter=100,                       
    cv=5,                             
    scoring='accuracy',               
    verbose=2,                        
    random_state=42,                  
    n_jobs=-1                         
)

# Realiza a busca com os dados de treino
random_search.fit(x_sample, y_sample)

# Exibe os melhores parâmetros encontrados
print("Melhores parâmetros encontrados:", random_search.best_params_)
print("Melhor desempenho:", random_search.best_score_)

# Selecionar cada parâmetro individualmente
best_n_estimators = random_search.best_params_['n_estimators']
best_max_depth = random_search.best_params_['max_depth']
```

```python
# Treinamento do modelo com os melhores parâmetros

# Criar o modelo Random Forest 
rf_clf = RandomForestClassifier(n_estimators= best_n_estimators, max_depth=best_max_depth,random_state=42)

# Treinar o modelo
rf_clf.fit(x_train, y_train)

# Fazer previsões
y_pred = rf_clf.predict(x_train)
```

```python
# Calculando as métricas de performance
random_forest_classifier_train_accuracy = mt.accuracy_score(y_train, y_pred)
random_forest_classifier_train_precision = mt.precision_score(y_train, y_pred, average='weighted')
random_forest_classifier_train_recall = mt.recall_score(y_train, y_pred, average='weighted')
random_forest_classifier_train_f1 = mt.f1_score(y_train, y_pred, average='weighted')

# Exibindo os resultados
print("Métricas de Performance:")
print(f"Accuracy: {random_forest_classifier_train_accuracy:.2f}")
print(f"Precision: {random_forest_classifier_train_precision:.2f}")
print(f"Recall: {random_forest_classifier_train_recall:.2f}")
print(f"F1-Score: {random_forest_classifier_train_f1:.2f}")
```
